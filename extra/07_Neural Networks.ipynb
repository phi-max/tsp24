# Neural Networks with Keras for classification
----

In this notebook, we will define our first neural network in `tensorflow` and `keras` and train it on the same data as in the first notebook. We start by importing some libraries:

```python
import numpy as np
from sklearn.datasets import load_iris
```

To limit GPU memory usage, we initialize TensorFlow first:

```python
import tensorflow as tf
```

Now load the keras modules:

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import np_utils
from keras.callbacks import TensorBoard
```

## Loading dataset
----
We start with the same dataset (Iris) as in the first example:

```python
iris = load_iris()
X = iris.data
y = iris.target
```

## Splitting into training and test data
----
Again, we have to split the data into training and test set:

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=0)
```

## One-hot encoding
----

The scikit-learn classifier accepts strings as different classes. For neural networks, the classes need to be one-hot encoded. This means that we have to generate a binary mask holding the ground truth labels.

For this, pre-defined functions exist:

```python
from keras.utils import to_categorical

train_y_ohe = to_categorical(train_y, 3)
test_y_ohe= to_categorical(test_y, 3)
```

Now we can look at what our data looks like hot-encoded:

```python
print(train_y[:3,])
```

```python
print(train_y_ohe[:3,])
```

## Defining a neural network in Keras
----
We use the sequential API in Keras and start by creating an instance of a keras model:

```python
model = Sequential()
```

The different layers of the network are added step by step using the add() function. We first define the size of the input layer.

How many input neurons do we need?


```python
INPUT_DIM = 4
```

We use a single dense layer with 16 neurons:

```python
model.add(Dense(16, input_shape=(INPUT_DIM,)))
```

We also have to specify the activation function for this layer. We use a sigmoid function:

```python
model.add(Activation('sigmoid'))
```

Next comes the output layer. How many output neurons do we need?


```python
OUT_N = 3
```

Define the fully connected output layer:

```python
model.add(Dense(OUT_N))
```

The activation function for the output layer is a softmax function:

```python
model.add(Activation('softmax'))
```

We can look at a summary of the network architecture and the number of parameters:

```python
model.summary()
```

## Compile model and specify metaparameters

Before we can fit the model to the data, we have to compile it to generate the corresponding TensorFlow graph that runs either on the CPU or GPU:

```python
model.compile(optimizer='rmsprop', 
              loss='categorical_crossentropy',
              metrics=["accuracy"])
```

## Training the neural network
----
The training process can be started using the fit() function, as in scikit-learn. We have to specify the one-hot encoded training data, the number of epochs, batch size, and the validation data.

First, we specify and create a log dir to later look at the training process with TensorBoard:

```python
import os

logdir = 'keras-nn'

if not os.path.exists(logdir): os.makedirs(logdir)
```

Now we specify number of epochs and batch size:

```python
epochs = 500
batch_size = 4
```

Finally, we train (fit) the neural network. This has to be repeated every time you change the parameters:

```python
history = model.fit(train_X, train_y_ohe, 
                    epochs=epochs, batch_size=batch_size,
                    validation_data=(test_X, test_y_ohe),
                    verbose=1,
                    callbacks=[TensorBoard(log_dir=logdir, histogram_freq=50)])
```

Note that if you repeat the training process without recompiling the model, it will continue from the previous state of weights.

## Evaluate network performance
----
We extract loss and accuracy using the evaluate() function. We are using the same test data as for training: 

```python
loss, accuracy = model.evaluate(test_X, test_y_ohe, verbose=0)
print("Accuracy = {:.2f}".format(accuracy))
```
